# Data Analysis and Visualizations

```python
canceled_percent = hotel_booking_df['is_canceled'].value_counts()
print(canceled_percent)

plt.figure(figsize = (3,4))
plt.title('Reservation status count')
plt.xlabel('status')
plt.ylabel('number of reservations')
plt.bar(['Not canceled','canceled'],hotel_booking_df['is_canceled'].value_counts(), edgecolor = 'k',width = 0.5)
plt.show()
```

```python
plt.figure(figsize = (8,5))
ax1= sns.countplot(x='hotel',hue = 'is_canceled',data = hotel_booking_df, palette = 'Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
for p in ax1.patches:
    if p.get_height() != 0:
        ax1.annotate(f'{p.get_height():,.2f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', fontsize=9, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.title('Reservation status in different hotels',size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
```

```python
resort_hotel = hotel_booking_df[hotel_booking_df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = hotel_booking_df[hotel_booking_df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts()
```

```python
city_hotel = hotel_booking_df[hotel_booking_df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel_adr = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel_adr = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (25,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 20)
plt.plot(resort_hotel_adr.index,resort_hotel_adr['adr'], label = 'Resort Hotel')
plt.plot(city_hotel_adr.index,city_hotel_adr['adr'], label = 'City Hotel')
plt.legend(fontsize = 10)
plt.show()
```

```python
hotel_booking_df['month'] =hotel_booking_df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month',hue = 'is_canceled',data = hotel_booking_df,palette = 'bright')
legend_labels,_ = ax1. get_legend_handles_labels()
for p in ax1.patches:
    if p.get_height() != 0:
        ax1.annotate(f'{p.get_height():}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', fontsize=7, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.title('Reservation Status per Month',size = 15)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
data=hotel_booking_df[hotel_booking_df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()
plt.figure(figsize = (12,6))
plt.title('ADR per Month', fontsize = 20)
ax1=sns.barplot(data =data ,x='month',y='adr' )
for p in ax1.patches:
    if p.get_height() != 0:
        ax1.annotate(f'{p.get_height():,.2f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', fontsize=7, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.legend(fontsize = 10)
plt.xlabel('Month')
plt.ylabel('Average Daily Rate')
plt.show()
```

```python
canceled_data = hotel_booking_df[hotel_booking_df['is_canceled'] == 1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct = '%.2f',labels = top_10_country.index)
plt.show()
```

```python
canceled_data = hotel_booking_df[hotel_booking_df['is_canceled'] == 1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
fig, ax = plt.subplots()

# Plot outer pie chart (larger)
outer_colors = plt.cm.tab10.colors[:len(top_10_country)]  # Use default color cycle
ax.pie(top_10_country, labels=top_10_country.index, autopct='%0.2f', startangle=10, pctdistance=.75, colors=outer_colors)

# Draw a circle to create the donut hole
centre_circle = plt.Circle((0,0),0.45,fc='white')
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal') 
plt.title('Top 10 countries with reservation canceled')

plt.show()
```
